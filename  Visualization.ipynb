```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import mplcursors
```

```python
# Directory where the CSV files are located
directory = './result_profitability/'

# Prepare an empty DataFrame for storing the results
results_df = pd.DataFrame()

# List all files in the directory
for filename in os.listdir(directory):
    if filename.endswith('_profitability.csv'):
        # Extract the date from the filename
        date = filename.split('_')[0]
        # Read the specific columns from the CSV file
        data = pd.read_csv(os.path.join(directory, filename), usecols=['Profit', 'Profitability Percent'])
        # Store the data in the DataFrame using the date as the index
        results_df.loc[date, 'Profit'] = data['Profit'].iloc[0]
        results_df.loc[date, 'Profitability Percent'] = data['Profitability Percent'].iloc[0]

# Convert the index to datetime format for proper sorting
results_df.index = pd.to_datetime(results_df.index)
# Sort the DataFrame by the index
results_df.sort_index(inplace=True)

```

```python
print(results_df)
```

```python
# Plot for Profit
plt.figure(figsize=(28, 7))

# Plot the 'Profit' data
plt.plot(results_df.index, results_df['Profit'], marker='o', linestyle='-', color='green')
plt.title('Daily Profit Over Time')
plt.xlabel('Date')
plt.ylabel('Profit')
plt.xticks(rotation=45)

# Improve the aesthetics for the x-axis
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator())
plt.grid(True)


# Ensure layout is tight so x-axis labels don't get cut off
plt.tight_layout()

# Show the plot
plt.show()
```

```python


# 
plt.figure(figsize=(24,7))
plt.plot(results_df.index, results_df['Profitability Percent'], marker='o')

#  y 
plt.ylim(results_df['Profitability Percent'].min() - 1, results_df['Profitability Percent'].max() + 1)

# 
plt.title('Daily Profit Percentage')
plt.xlabel('Date')
plt.ylabel('Profit Percent (%)')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator())

#  x 
plt.xticks(rotation=45)

# 
plt.tight_layout()
plt.grid(True)

# 
plt.show()
```

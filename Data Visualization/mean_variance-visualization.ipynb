```python
import pandas as pd
import os

directory_path = 'C:\\Users\\zoec0\\Desktop\\DSMP\\Clean\\LOBs'

# Initialize a DataFrame to store results
results_df = pd.DataFrame(columns=['filename', 'mean_ask_price1', 'variance_ask_price1', 'mean_bid_price1', 'variance_bid_price1'])

for filename in os.listdir(directory_path):
    if filename.endswith('.txt'):  # Adjust for your file extension, e.g., '.csv'
        file_path = os.path.join(directory_path, filename)
        
        # Load the CSV file into a DataFrame
        df = pd.read_csv(file_path)
        
        # Extract the 'ask_price1' and 'bid_price1' columns
        ask_price1 = df['ask_price1']
        bid_price1 = df['bid_price1']
        
        # Calculate the mean and variance
        mean_ask_price1 = ask_price1.mean()
        variance_ask_price1 = ask_price1.var()
        mean_bid_price1 = bid_price1.mean()
        variance_bid_price1 = bid_price1.var()
        
        # Append the results for this file to the DataFrame
        results_df = results_df.append({
            'filename': filename,
            'mean_ask_price1': mean_ask_price1,
            'variance_ask_price1': variance_ask_price1,
            'mean_bid_price1': mean_bid_price1,
            'variance_bid_price1': variance_bid_price1
        }, ignore_index=True)

# Define the path for the output CSV file
output_file_path = 'C:\\Users\\zoec0\\Desktop\\DSMP\\LOB_means_variances.csv'

# Export the DataFrame to a CSV file
results_df.to_csv(output_file_path, index=False)

print(f'Results exported to {output_file_path}')
```

```python
import pandas as pd
import os

directory_path = 'C:\\Users\\zoec0\\Desktop\\DSMP\\Clean\\Tape'
results = []

for filename in os.listdir(directory_path):
    if filename.endswith('.csv'):
        file_path = os.path.join(directory_path, filename)
        
        date_part = filename.split('_')[3]  
        date_part = date_part.replace('tapes.csv', '')  
        
        df_tape = pd.read_csv(file_path)

        # Check if 'Weighted_Price' column exists
        if 'Weighted_Price' in df_tape.columns:
            weighted_price = df_tape['Weighted_Price']
            mean_weighted_price = weighted_price.mean()
            variance_weighted_price = weighted_price.var()
            results.append([filename, mean_weighted_price, variance_weighted_price])
        else:
            print(f"Column 'Weighted_Price' not found in {filename}. Skipping this file.")

# Convert results to DataFrame
results_df = pd.DataFrame(results, columns=['filename', 'mean_weighted_price', 'variance_weighted_price'])

# Define the output file path
output_file_path = 'C:\\Users\\zoec0\\Desktop\\DSMP\\Tapes_means_variances.csv'

# Export to CSV
results_df.to_csv(output_file_path, index=False)
print(f'Results exported to {output_file_path}')
```

```python
import pandas as pd
import os

directory_path = 'C:\\Users\\zoec0\\Desktop\\DSMP\\Clean\\LOBs'

# Initialize a DataFrame to store results, with 'date' instead of 'filename'
results_df = pd.DataFrame(columns=['date', 'mean_ask_price1', 'variance_ask_price1', 'mean_bid_price1', 'variance_bid_price1'])

for filename in os.listdir(directory_path):
    if filename.endswith('.txt'):  # Adjust for your file extension, e.g., '.csv'
        file_path = os.path.join(directory_path, filename)
        
        # Extract the date from the filename more precisely
        # Assuming the format is always "Clean_UoB_Set01_YYYY-MM-DDLOBs.txt"
        date_part = filename.split('_')[3]  # This gets '2025-01-02LOBs'
        date_part = date_part.replace('LOBs.txt', '')  # Now it's just '2025-01-02'
        
        # Load the CSV file into a DataFrame
        df = pd.read_csv(file_path)
        
        # Extract the 'ask_price1' and 'bid_price1' columns
        ask_price1 = df['ask_price1']
        bid_price1 = df['bid_price1']
        
        # Calculate the mean and variance
        mean_ask_price1 = ask_price1.mean()
        variance_ask_price1 = ask_price1.var()
        mean_bid_price1 = bid_price1.mean()
        variance_bid_price1 = bid_price1.var()
        
        # Append the results for this file to the DataFrame, using the correctly formatted date
        results_df = results_df.append({
            'date': date_part,
            'mean_ask_price1': mean_ask_price1,
            'variance_ask_price1': variance_ask_price1,
            'mean_bid_price1': mean_bid_price1,
            'variance_bid_price1': variance_bid_price1
        }, ignore_index=True)

# Define the path for the output CSV file
output_file_path = 'C:\\Users\\zoec0\\Desktop\\DSMP\\LOB_means_variances.csv'

# Export the DataFrame to a CSV file
results_df.to_csv(output_file_path, index=False)

print(f'Results exported to {output_file_path}')
```

```python
import pandas as pd
import os

directory_path = 'C:\\Users\\zoec0\\Desktop\\DSMP\\Clean\\Tape'
results = []

for filename in os.listdir(directory_path):
    if filename.endswith('.csv'):
        # Extract the date from the filename
        # Assuming the date format is always "YYYY-MM-DD"
        date = filename.split('_')[3].split('tapes.csv')[0]
        file_path = os.path.join(directory_path, filename)
        df_tape = pd.read_csv(file_path)

        # Check if 'Weighted_Price' column exists
        if 'Weighted_Price' in df_tape.columns:
            weighted_price = df_tape['Weighted_Price']
            mean_weighted_price = weighted_price.mean()
            variance_weighted_price = weighted_price.var()
            # Append the date instead of the filename to the results
            results.append([date, mean_weighted_price, variance_weighted_price])
        else:
            print(f"Column 'Weighted_Price' not found in {filename}. Skipping this file.")

# Convert results to DataFrame, now with the 'date' column correctly extracted
results_df = pd.DataFrame(results, columns=['date', 'mean_weighted_price', 'variance_weighted_price'])

# Define the output file path
output_file_path = 'C:\\Users\\zoec0\\Desktop\\DSMP\\Tapes_means_variances.csv'

# Export to CSV
results_df.to_csv(output_file_path, index=False)
print(f'Results exported to {output_file_path}')
```
